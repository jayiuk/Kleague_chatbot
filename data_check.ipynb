```python
def get_concat(path):
    data = pd.DataFrame()
    list = os.listdir(path)
    for i in range(0, len(list)):
        df = pd.read_csv(os.path.join(path, list[i]))
        data = pd.concat([data, df])
    return data
```

```python
path1 = './data/jt_mt'
path2 = './data/jt_por'
path3 = './data/jt_wk'
path4 = './data/jt_por_info'
jt_mt_df = get_concat(path1)
jt_por_df = get_concat(path2)
jt_wk_df = get_concat(path3)
jt_por_info_df = get_concat(path4)
```
